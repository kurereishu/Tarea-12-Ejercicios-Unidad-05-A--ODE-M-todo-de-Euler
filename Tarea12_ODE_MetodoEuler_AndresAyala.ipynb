```python
import numpy as np
import matplotlib.pyplot as plt

def euler_method(f, t0, y0, h, n):
    """
    Método de Euler para resolver ecuaciones diferenciales ordinarias
    
    Parámetros:
    f: función f(t, y) que define la EDO y' = f(t, y)
    t0: valor inicial de t
    y0: valor inicial de y
    h: tamaño del paso
    n: número de pasos
    
    Retorna:
    t: array de valores de t
    y: array de valores aproximados de y
    """
    t = np.zeros(n+1)
    y = np.zeros(n+1)
    t[0] = t0
    y[0] = y0
    
    for i in range(n):
        y[i+1] = y[i] + h * f(t[i], y[i])
        t[i+1] = t[i] + h
    
    return t, y

# Definición de los problemas

# Problema a: y' = t*e^(3t) - 2y
def problema_a():
    f = lambda t, y: t * np.exp(3*t) - 2*y
    t0, y0 = 0, 0
    h = 0.5
    n = int((1 - t0)/h)
    
    t, y = euler_method(f, t0, y0, h, n)
    
    print("Problema a: y' = t*e^(3t) - 2y, y(0)=0, h=0.5")
    for i in range(len(t)):
        print(f"t = {t[i]:.1f}: y = {y[i]:.6f}")
    
    # Gráfica
    plt.figure(figsize=(8, 4))
    plt.plot(t, y, 'bo-', label='Aproximación Euler')
    plt.title("Problema a: Solución aproximada con método de Euler")
    plt.xlabel('t')
    plt.ylabel('y(t)')
    plt.legend()
    plt.grid()
    plt.show()

# Problema b: y' = 1 + (t - y)^2
def problema_b():
    f = lambda t, y: 1 + (t - y)**2
    t0, y0 = 2, 1
    h = 0.5
    n = int((3 - t0)/h)
    
    t, y = euler_method(f, t0, y0, h, n)
    
    print("\nProblema b: y' = 1 + (t - y)^2, y(2)=1, h=0.5")
    for i in range(len(t)):
        print(f"t = {t[i]:.1f}: y = {y[i]:.6f}")
    
    # Gráfica
    plt.figure(figsize=(8, 4))
    plt.plot(t, y, 'ro-', label='Aproximación Euler')
    plt.title("Problema b: Solución aproximada con método de Euler")
    plt.xlabel('t')
    plt.ylabel('y(t)')
    plt.legend()
    plt.grid()
    plt.show()

# Problema c: y' = 1 + 1/t
def problema_c():
    f = lambda t, y: 1 + 1/t
    t0, y0 = 1, 2
    h = 0.25
    n = int((2 - t0)/h)
    
    t, y = euler_method(f, t0, y0, h, n)
    
    print("\nProblema c: y' = 1 + 1/t, y(1)=2, h=0.25")
    for i in range(len(t)):
        print(f"t = {t[i]:.2f}: y = {y[i]:.6f}")
    
    # Gráfica
    plt.figure(figsize=(8, 4))
    plt.plot(t, y, 'go-', label='Aproximación Euler')
    plt.title("Problema c: Solución aproximada con método de Euler")
    plt.xlabel('t')
    plt.ylabel('y(t)')
    plt.legend()
    plt.grid()
    plt.show()

# Problema d: y' = cos(2t) + sin(3t)
def problema_d():
    f = lambda t, y: np.cos(2*t) + np.sin(3*t)
    t0, y0 = 0, 1
    h = 0.25
    n = int((1 - t0)/h)
    
    t, y = euler_method(f, t0, y0, h, n)
    
    print("\nProblema d: y' = cos(2t) + sin(3t), y(0)=1, h=0.25")
    for i in range(len(t)):
        print(f"t = {t[i]:.2f}: y = {y[i]:.6f}")
    
    # Gráfica
    plt.figure(figsize=(8, 4))
    plt.plot(t, y, 'mo-', label='Aproximación Euler')
    plt.title("Problema d: Solución aproximada con método de Euler")
    plt.xlabel('t')
    plt.ylabel('y(t)')
    plt.legend()
    plt.grid()
    plt.show()

# Ejecutar todos los problemas
if __name__ == "__main__":
    problema_a()
    problema_b()
    problema_c()
    problema_d()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import exp, cos, sin, log

## =============================================
## Métodos de Taylor implementados
## =============================================

def taylor_order2(f, df, t0, y0, h, n):
    """Método de Taylor de orden 2 para resolver EDOs"""
    t = np.zeros(n+1)
    y = np.zeros(n+1)
    t[0] = t0
    y[0] = y0
    
    for i in range(n):
        y[i+1] = y[i] + h * f(t[i], y[i]) + (h**2)/2 * df(t[i], y[i])
        t[i+1] = t[i] + h
    
    return t, y

def taylor_order4(f, df, d2f, d3f, t0, y0, h, n):
    """Método de Taylor de orden 4 para resolver EDOs"""
    t = np.zeros(n+1)
    y = np.zeros(n+1)
    t[0] = t0
    y[0] = y0
    
    for i in range(n):
        ti, yi = t[i], y[i]
        h1 = h * f(ti, yi)
        h2 = h * df(ti, yi)
        h3 = h * d2f(ti, yi)
        h4 = h * d3f(ti, yi)
        
        y[i+1] = yi + h1 + (h**2)/2 * h2 + (h**3)/6 * h3 + (h**4)/24 * h4
        t[i+1] = ti + h
    
    return t, y

## =============================================
## Soluciones exactas para comparación
## =============================================

exact_solutions = {
    'a': lambda t: (1/5)*t*exp(3*t) - (1/25)*exp(3*t) + (1/25)*exp(-2*t),
    'b': lambda t: t + 1/(1-t),
    'c': lambda t: t*log(t) + 2*t if t > 0 else 0,
    'd': lambda t: (1/2)*sin(2*t) - (1/3)*cos(3*t) + 4/3
}

## =============================================
## Derivadas necesarias para los métodos de Taylor
## =============================================

# Problema 6a/7a: y' = te^(3t) - 2y
f6a = lambda t, y: t * exp(3*t) - 2*y
df6a = lambda t, y: exp(3*t) + 3*t*exp(3*t) - 2*f6a(t, y)
d2f6a = lambda t, y: 6*exp(3*t) + 9*t*exp(3*t) - 2*df6a(t, y)
d3f6a = lambda t, y: 27*exp(3*t) + 27*t*exp(3*t) - 2*d2f6a(t, y)

# Problema 6b/7b: y' = 1 + (t - y)^2
f6b = lambda t, y: 1 + (t - y)**2
df6b = lambda t, y: 2*(t - y)*(1 - f6b(t, y))
d2f6b = lambda t, y: 2*(1 - f6b(t, y))**2 - 2*(t - y)*df6b(t, y)
d3f6b = lambda t, y: -6*(1 - f6b(t, y))*df6b(t, y) - 2*(t - y)*d2f6b(t, y)

# Problema 6c/7c: y' = 1 + y/t
f6c = lambda t, y: 1 + y/t if t != 0 else np.nan
df6c = lambda t, y: -y/t**2 + f6c(t, y)/t if t != 0 else np.nan
d2f6c = lambda t, y: (2*y)/t**3 - (2*f6c(t, y))/t**2 + df6c(t, y)/t if t != 0 else np.nan
d3f6c = lambda t, y: (-6*y)/t**4 + (6*f6c(t, y))/t**3 - (3*df6c(t, y))/t**2 + d2f6c(t, y)/t if t != 0 else np.nan

# Problema 6d/7d: y' = cos(2t) + sin(3t)
f6d = lambda t, y: cos(2*t) + sin(3*t)
df6d = lambda t, y: -2*sin(2*t) + 3*cos(3*t)
d2f6d = lambda t, y: -4*cos(2*t) - 9*sin(3*t)
d3f6d = lambda t, y: 8*sin(2*t) - 27*cos(3*t)

## =============================================
## Ejercicio 6: Método de Taylor orden 2
## =============================================

def ejercicio_6():
    print("\n" + "="*60)
    print("EJERCICIO 6: MÉTODO DE TAYLOR ORDEN 2")
    print("="*60)
    
    # Problema 6a
    print("\nProblema 6a: y' = te^(3t) - 2y, y(0)=0, h=0.5")
    t0, y0 = 0, 0
    h = 0.5
    n = int(1/h)
    t, y = taylor_order2(f6a, df6a, t0, y0, h, n)
    for i in range(len(t)):
        exact = exact_solutions['a'](t[i])
        print(f"t = {t[i]:.1f}: y = {y[i]:.6f}, Exacta = {exact:.6f}, Error = {abs(exact - y[i]):.6f}")
    
    # Problema 6b
    print("\nProblema 6b: y' = 1 + (t-y)^2, y(2)=1, h=0.5")
    t0, y0 = 2, 1
    h = 0.5
    n = int((3 - t0)/h)
    t, y = taylor_order2(f6b, df6b, t0, y0, h, n)
    for i in range(len(t)):
        exact = exact_solutions['b'](t[i])
        print(f"t = {t[i]:.1f}: y = {y[i]:.6f}, Exacta = {exact:.6f}, Error = {abs(exact - y[i]):.6f}")
    
    # Problema 6c
    print("\nProblema 6c: y' = 1 + y/t, y(1)=2, h=0.25")
    t0, y0 = 1, 2
    h = 0.25
    n = int((2 - t0)/h)
    t, y = taylor_order2(f6c, df6c, t0, y0, h, n)
    for i in range(len(t)):
        exact = exact_solutions['c'](t[i])
        print(f"t = {t[i]:.2f}: y = {y[i]:.6f}, Exacta = {exact:.6f}, Error = {abs(exact - y[i]):.6f}")
    
    # Problema 6d
    print("\nProblema 6d: y' = cos(2t) + sin(3t), y(0)=1, h=0.25")
    t0, y0 = 0, 1
    h = 0.25
    n = int(1/h)
    t, y = taylor_order2(f6d, df6d, t0, y0, h, n)
    for i in range(len(t)):
        exact = exact_solutions['d'](t[i])
        print(f"t = {t[i]:.2f}: y = {y[i]:.6f}, Exacta = {exact:.6f}, Error = {abs(exact - y[i]):.6f}")

## =============================================
## Ejercicio 7: Método de Taylor orden 4
## =============================================

def ejercicio_7():
    print("\n" + "="*60)
    print("EJERCICIO 7: MÉTODO DE TAYLOR ORDEN 4")
    print("="*60)
    
    # Problema 7a
    print("\nProblema 7a: y' = te^(3t) - 2y, y(0)=0, h=0.5")
    t0, y0 = 0, 0
    h = 0.5
    n = int(1/h)
    t, y = taylor_order4(f6a, df6a, d2f6a, d3f6a, t0, y0, h, n)
    for i in range(len(t)):
        exact = exact_solutions['a'](t[i])
        print(f"t = {t[i]:.1f}: y = {y[i]:.6f}, Exacta = {exact:.6f}, Error = {abs(exact - y[i]):.6f}")
    
    # Problema 7b
    print("\nProblema 7b: y' = 1 + (t-y)^2, y(2)=1, h=0.5")
    t0, y0 = 2, 1
    h = 0.5
    n = int((3 - t0)/h)
    t, y = taylor_order4(f6b, df6b, d2f6b, d3f6b, t0, y0, h, n)
    for i in range(len(t)):
        exact = exact_solutions['b'](t[i])
        print(f"t = {t[i]:.1f}: y = {y[i]:.6f}, Exacta = {exact:.6f}, Error = {abs(exact - y[i]):.6f}")
    
    # Problema 7c
    print("\nProblema 7c: y' = 1 + y/t, y(1)=2, h=0.25")
    t0, y0 = 1, 2
    h = 0.25
    n = int((2 - t0)/h)
    t, y = taylor_order4(f6c, df6c, d2f6c, d3f6c, t0, y0, h, n)
    for i in range(len(t)):
        exact = exact_solutions['c'](t[i])
        print(f"t = {t[i]:.2f}: y = {y[i]:.6f}, Exacta = {exact:.6f}, Error = {abs(exact - y[i]):.6f}")
    
    # Problema 7d
    print("\nProblema 7d: y' = cos(2t) + sin(3t), y(0)=1, h=0.25")
    t0, y0 = 0, 1
    h = 0.25
    n = int(1/h)
    t, y = taylor_order4(f6d, df6d, d2f6d, d3f6d, t0, y0, h, n)
    for i in range(len(t)):
        exact = exact_solutions['d'](t[i])
        print(f"t = {t[i]:.2f}: y = {y[i]:.6f}, Exacta = {exact:.6f}, Error = {abs(exact - y[i]):.6f}")

## =============================================
## Ejecutar todos los ejercicios
## =============================================

if __name__ == "__main__":
    ejercicio_6()  # Ejercicio 6: Taylor orden 2
    
```
